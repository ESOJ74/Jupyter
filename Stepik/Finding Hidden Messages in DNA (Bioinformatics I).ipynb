```python
def PatternCount(Text, Pattern):
    count = 0
    for i in range(len(Text)-len(Pattern)+1):
        if Text[i:i+len(Pattern)] == Pattern:
            count = count+1
    return count 
```

## Desafio 2

Code Challenge: Solve the Frequent Words Problem.

Input: A string Text and an integer k.

Output: All most frequent k-mers in Text.

```python
def frequencyMap(text, k):
    freq = {}
    n = len(text)
    for i in range(n-k+1):
        pattern = text[i:i+k]     
        if pattern in freq:
            freq[pattern] = freq[pattern]+1
        else:
            freq.setdefault(pattern, 1)
    return freq  
```

```python
f = open ('dataset_2_10.txt')
frequencyMap(f.read(),12)
```

```python
def frequentWords(text, k):
    words = []
    freq = frequencyMap(text, k)
    m = max(freq.values())
    for key in freq:
        if freq[key] == m:
            words.append(key)
    return words
```

```python
f = open ('dataset_2_10.txt')
g = FrequentWords(f.read(),12)
while g:
    print(g.pop())
```

```python
def ReverseComplement(pattern):
    pattern = complement(pattern) # complement each letter in a string
    return pattern

def complement(pattern):
    complementPattern = ""
    for char in pattern:
        if char == "A":
            complementPattern = "T" + complementPattern;
        if char == "T": 
            complementPattern = "A" + complementPattern;
        if char == "G":
            complementPattern = "C" + complementPattern;
        if char == "C":
            complementPattern = "G" + complementPattern;
    return complementPattern
```
